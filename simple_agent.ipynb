```python
from langchain_groq import ChatGroq
from utils import load_md_files, generate_system_prompt, save_memory, load_memory
from dotenv import load_dotenv
load_dotenv()
from langchain.chat_models import init_chat_model
import json

# llm = ChatGroq(
#     model="openai/gpt-oss-120b",
#     api_key="YOUR_GROQ_API_KEY"
# )
llm = init_chat_model(
  model="openai/gpt-oss-120b",
  model_provider="groq", 
)
# Load all markdown files
md_files = load_md_files("raw")

system_prompt = generate_system_prompt(md_files)

memory_path = "memories/simple/chat.json"
memory = load_memory(memory_path)


def chat(user_question: str):

    messages = []

    messages.append({"role": "system", "content": system_prompt})

    messages.extend(memory)

    messages.append({"role": "user", "content": user_question})

    response = llm.invoke(messages)

    memory.append({"role": "user", "content": user_question})
    memory.append({"role": "assistant", "content": response.content})

    save_memory(memory_path, memory)

    return response.content


# Example
print(chat("What is this knowledge base about?"))
```
